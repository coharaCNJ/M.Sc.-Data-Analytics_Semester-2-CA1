```python
import os
from tqdm import tqdm
import PIL.Image
import numpy as np #For numerical functions, incl. statistical.
import matplotlib.pyplot as plt #2D plotting library
import cv2
from keras.utils import np_utils
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Conv2D, Flatten, Dense, MaxPool2D
import tensorflow as tf
```

```python
classes = {'Mayflies':0,'Stoneflies':1,'Caddis_flies':2,'Waterlouse':3}
```

```python
mayfly_directory = os.getcwd() + '/Species_images/Mayflies/'
stonefly_directory = os.getcwd() + '/Species_images/Stoneflies'
caddis_directory = os.getcwd() + '/Species_images/Trichoptera (caddis fly)'
Waterlouse_directory = os.getcwd() + '/Species_images/Asellus aquaticus  (waterlouse)'
```

```python
def assign_label(img,genus):
    return genus

X = []
y = []

def make_data(genus,directory):
    print(genus)
    print(directory)
    for img in tqdm(os.listdir(directory)):
        label=assign_label(img,genus)
        path = os.path.join(directory,img)
        print(path)
        img = PIL.Image.open(path)
        #img = cv2.resize(img, (IMAGE_WIDTH,IMAGE_HEIGHT))
        
        X.append(np.array(img))
        y.append(str(label))
```

```python
make_data(classes.get('Mayflies'), mayfly_directory)
make_data(classes.get('Stoneflies'), stonefly_directory)
make_data(classes.get('Caddis_flies'), caddis_directory)
make_data(classes.get('Waterlouse'), Waterlouse_directory)
```

```python
X = np.array(X)
X /= 255
X.shape
```

```python
min_width = 0
min_height = 0

for x in range(len(X)):
    min_width = X[x].shape[1]
    min_height = X[x].shape[0]
    if X[x].shape[1] < min_width:
        min_width = X[x].shape[1]
    if X[x].shape[0] < min_height:
        min_height = X[x].shape[0]
```

```python
image = X[3]

plt.imshow(image)
```

```python
#resize the images to feed the model
for x in range(len(X)):
    X[x] = cv2.resize(X[x], (min_width, min_height)) 
```

```python
#Convert the list to an array:
y = np.array(y)

#The number of species = 4. One hot encode these:
n_classes = 4
print("Shape before one-hot encoding: ", y.shape)
y = np_utils.to_categorical(y, n_classes)
print("Shape after one-hot encoding: ", y.shape)
```

```python
model = Sequential()
# convolutional layer (a layer that applies filters - kernel size is the size of the filter (a Sobol 3x3 filter),
#Strides is how far the filter moves each time, in this case it moves 1 pixel at a time)

model.add(Conv2D(25, kernel_size=(3,3), strides=(1,1), padding='valid', activation='relu',input_shape=(X[0].shape[0],X[0].shape[1],3)))
model.add(MaxPool2D(pool_size=(1,1))) #A filter that takes the max pixel value that falls within it
# flatten output of conv
model.add(Flatten())
# hidden layer
model.add(Dense(100, activation='relu')) #One layer neural network
# output layer
model.add(Dense(4, activation='softmax')) #Softmax because it is a classification problem
```

```python
# compiling the sequential model
model.compile(loss='categorical_crossentropy', metrics=['accuracy'], optimizer='adam')
```

```python
#model.fit is throwing an error: 'ValueError: Failed to convert a NumPy array to a Tensor (Unsupported object type numpy.ndarray).'
#Need to reshape the images file.
images = []
for x in range(len(X)):
    images.append(X[x])
```

```python
images = np.array(images) 
```

```python
images.shape
```

```python
images = images.astype('float32')
```

```python
print(images.dtype), print(y.dtype)
```

```python
images = tf.convert_to_tensor(images) 
```

```python
# training the model
model.fit(images, y, batch_size=4, epochs=10, validation_data=(images, y))
```
