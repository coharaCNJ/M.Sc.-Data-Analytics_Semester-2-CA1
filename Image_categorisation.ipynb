```python
def assign_label(img,genus):
    return genus

X = []
y = []

def make_data(genus,directory):
    print(genus)
    print(directory)
    for img in tqdm(os.listdir(directory)):
        label=assign_label(img,genus)
        path = os.path.join(directory,img)
        print(path)
        img = PIL.Image.open(path)
        #img = cv2.resize(img, (IMAGE_WIDTH,IMAGE_HEIGHT))
        
        X.append(np.array(img))
        y.append(str(label))
```

```python
print(classes.get('Mayflies'))
```

```python
make_data(classes.get('Mayflies'), mayfly_directory)
make_data(classes.get('Stoneflies'), stonefly_directory)
make_data(classes.get('Caddis_flies'), caddis_directory)
make_data(classes.get('Waterlouse'), Waterlouse_directory)
```
